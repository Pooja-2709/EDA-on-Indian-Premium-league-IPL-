# Merge the dataset

```python
merge = pd.merge(delivery,match, left_on='match_id', right_on='id')
merge.head()
```

```python
merge.describe()
```

# Batsman with highest no. of sixes

```python
sixes = merge[merge['batsman_runs']==6]
c=sixes.groupby('batsman').count()['batsman_runs'].sort_values(ascending=False).head()
c
```

```python
plt.figure(figsize = (10,5))
sns.barplot(data = merge, y=c, x=c.index)
```

# Top wicket takers

```python
delivery.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head().plot(kind='bar', color='blue')
plt.xlabel('Bowler')
plt.ylabel('Total wickets taken')
plt.show()
```
